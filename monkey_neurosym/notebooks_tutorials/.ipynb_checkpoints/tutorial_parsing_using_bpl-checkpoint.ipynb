## ENTIRE PARSING PIPELINE HERE

Note: Below is the the pipeline for parsing except not including GNS of pyBPL models. just taking best K parses based on a cost function for distance traveled (or ink). 

If you want to integrate with GNS parsing, then is simple. just look into get_parses_from_strokes and use the part that goes from strokes to image. Then take that image and pass into the GNS as you would the omniglot images.

If want to use "extra junctions" (which is useful if you are passing in stimulus images, as opposed to monkey beahviro images) then need to do one additional thing, which is to pass
in extra_junctions=True to generate_random_parses, which is is get_topK_parses (again, see within get_parses_from_strokes). THis will preprocess to pull out corners, passing into generate_random_parses, which otherwise will run same as the GNS version
                                                       

```python
from pythonlib.drawmodel.parsing import get_parses_from_strokes, score_function

if stim_ver=="task_image":
    use_extra_junctions=True
elif stim_ver=="monkey_strokes":
    use_extra_junctions=False

score_fn = lambda parses: score_function(parses, ver="travel", 
                                         normalization="negative", use_torch=True)
parses_k, log_probs_k = get_parses_from_strokes(strokes, canvas_max_WH, image_WH,
                                               use_extra_junctions=use_extra_junctions, 
                                                plot=True)
        
parses = parses_k
    
```
